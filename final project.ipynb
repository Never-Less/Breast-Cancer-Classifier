# Data Preprocessing

1. Change diagnosis Malignant as 1 and Benign as 0
2. Remove some of the outlier using Z-score method

```python
#Changing diagnosis label as 0 for benign and 1 for malignant

df_process = df
df_process['diagnosis'] = df['diagnosis'].replace(['B', 'M'], [0, 1])
```

```python
df_process.sample(10)
```

```python
df_process['diagnosis'].dtypes
```

```python
#Removing outlier using Z-score method
from scipy.stats import zscore

z = np.abs(zscore(df_process[df_process.columns]))
df_clear=df_process[(z<3.0).all(axis=1)]
```

```python
print(df_process.shape[0])
print(df_clear.shape[0])
```

```python
# Split data train and data test

from sklearn.model_selection import train_test_split

data_columns = df_clear.columns.values.tolist()
y = ['diagnosis']
x = [column for column in data_columns if column not in y]
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_clear[x], df_clear[y], test_size = 0.2, random_state = 19)
print(x_train.shape)
print(x_test.shape)
```

```python
# Normalize data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
scaler.transform(x_train)
scaler.transform(x_test)
```

```python
# Feature selection using correlation
fig = plt.subplots(figsize=(15,15))
sns.heatmap(df_clear.corr(), annot=True)
```

```python
corr_target = abs(df_clear.corr()['diagnosis'])
relevant_features = corr_target[(corr_target > 0.2) & (corr_target < 0.8)]
print(relevant_features.index, len(relevant_features))
```

```python
x_train = x_train[x_train.columns.intersection(relevant_features.index)]
x_test = x_test[x_test.columns.intersection(relevant_features.index)]
```

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
score = model.score(x_test, y_test)
print("the accuracy is: %.2f" % (score * 100))
```

# F1 score of Logistic Regression model

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cf_matrix = confusion_matrix(y_test, y_pred)

group_names = ["True Negative", "False Positive", "False Negative", "True Positive"]
group_counts = ["{0:0.0f}".format(value) for value in cf_matrix.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in cf_matrix.flatten()/np.sum(cf_matrix)]

labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)

plt.figure(figsize=(10, 8))
sns.heatmap(cf_matrix, annot=labels, fmt='', cmap='Blues')
```
